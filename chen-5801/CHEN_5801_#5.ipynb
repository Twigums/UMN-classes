# Problem 2

a. 

For the temperature approach constraint, the disjunction is given by: $$\begin{bmatrix}
x \\
T_{in} - T_{out} \ge \Delta T_{min}
\end{bmatrix} \lor \begin{bmatrix}
\neg x \\
T_{in} - T_{out} = \Delta T_{min} = 0
\end{bmatrix},$$

since if there is not a heat exchanger, the temperature between those "two points" should be equal.

For the disjunction: $T_{in} - T_{out} \ge \Delta T_{min}$, we can first flip the inequality to: $-(T_{in} - T_{out}) \le -\Delta T_{min}$. Then, the big-M formulation is given by: $-(T_{in} - T_{out}) \le -\Delta T_{min} + M (1 - y)$, where $y \in \{0, 1\}$ is the binary variable representing the choice of the heat exchanger. For $T_{in} - T_{out} = 0,$ we can bound $T_{in} - T_{out} \le 0, T_{in} - T_{out} \ge 0$. So, we will have: $T_{in} - T_{out} \le M (1 - y), -(T_{in} - T_{out}) \le M (1 - y)$.

For convex hull, assume $\vec{T} = \begin{bmatrix}
T_{in} \\
T_{out}
\end{bmatrix} = z_1 + z_2, z_1 = \begin{bmatrix}
v_1 \\
w_1
\end{bmatrix}, z_2 = \begin{bmatrix}
v_2 \\
w_2
\end{bmatrix}.$

Then,

$T_{in} = v_1 + v_2,$

$T_{out} = w_1 + w_2,$

$-v_1 + w_1 \le -\Delta T_{min} y,$

$v_2 - w_2 = 0,$

$0 \le v_1 \le \Delta T_{min} y,$

$0 \le w_1 \le \Delta T_{min} y,$

$y \in \{0, 1\}.$

# Problem 3

a.

Assume on an hour basis.

We want to maximize profit, which is the money made from selling C minus cost of purchase and operating costs:

$\max \: c_C - (c_A + c_B + \sum_{i = 1} ^ 3 c_i y_i),$

where $c_C = 1800 m_C$, and $m_C = y_2 (0.82m_{B, 2}) + y_3 (0.95m_{B, 3}),$

$c_A = 500 m_A$, where $m_A$ is the mass purchased.

$c_B = 950 m_B$, where $m_B = y_1 m_{B, 1} - (y_2 m_{B, 2} + y_3 m_{B, 3})$, where $m_B$ is how much of B we should purchase to have in stock along with what was made from I to operate II and III.

$m_{B, 1} = 0.9m_A$

This can be rewritten as such:

$\max \: c_C - (c_A + c_B + c_1 + c_2 + c_3)$

$\text{s.t.} \: \: 0.82m_{B, 2} \le 10 * y_2$ (if using one process, the amount produced should be less than demand; otherwise, 0)

$\: \: \: \: \: \: \: 0.95m_{B, 3} \le 10 * y_3$

$\: \: \: \: \: \: \: c_A = 500 m_A$

$\: \: \: \: \: \: \: c_B = 950 m_B$

$\: \: \: \: \: \: \: c_C = 1800 m_C$

$\: \: \: \: \: \: \: c_1 = 1000 * y_1 + 250 * m_A$

$\: \: \: \: \: \: \: c_2 = 1500 * y_2 + 400 * m_{B, 2}$

$\: \: \: \: \: \: \: c_3 = 2000 * y_3 + 550 * m_{B, 3}$

$\: \: \: \: \: \: \: m_C = 0.82m_{B, 2} + 0.95m_{B, 3}$

$\: \: \: \: \: \: \: y_2 + y_3 = 1$ (exclusivity)

$\: \: \: \: \: \: \: 0 \le m_A \le 16, m_{B, 2} \ge 0, m_{B,3} \ge 0$

$\: \: \: \: \: \: \: y_i \in \{0, 1\} \: \forall i \in [1, 3]$

```julia
# part a

using JuMP, GAMS

function plant_a()
    
    model = Model(GAMS.Optimizer)
    set_optimizer_attribute(model, "OptCR", 0)
    
    @variables(model, begin
        y1, Bin
        y2, Bin
        y3, Bin
        m_A >= 0
        m_C >= 0
        m_B0 >= 0
        m_B1 >= 0
        m_B2 >= 0
        m_B3 >= 0
        
    end)
    @constraints(model, begin
        process_one, m_B1 == 0.9 * m_A
        B_total, m_B0 + m_B1 == m_B2 + m_B3 # buy + made = used by 2 + used by 3
        C_total, m_C == 0.82 * m_B2 + 0.95 * m_B3 # total amount produced = conversion by 2 + conversion by 3
        demand_C2, 0.82 * m_B2 <= 10 * y2 # mass of B in process 2 should be 0 if off; otherwise, C it produces should be lower than demand
        demand_C3, 0.95 * m_B3 <= 10 * y3 # same as above
        demand_A, m_A <= 16 # demand of A
        exclusivity, y2 + y3 == 1 # we either use process 2 or 3
        
    end)
    @objective(model, Max, 1800 * (m_C) - (500 * m_A + 950 * m_B1 + (1000 * y1 + 250 * m_A) + (1500 * y2 + 400 * m_B2) + (2000 * y3 + 550 * m_B3)))
    
    print(model)
    optimize!(model)
    println("objective: ", objective_value(model))
    println("y1: ", value.(y1))
    println("y2: ", value.(y2))
    println("y3: ", value.(y3))
    println("m_A: ", value.(m_A))
    println("m_B0: ", value.(m_B1))                    
    println("m_B1: ", value.(m_B1))
    println("m_B2: ", value.(m_B2))
    println("m_B3: ", value.(m_B3))
    
end
```

```julia
plant_a()
```

```julia
# part b
# for piecewise, we can split C into C1 and C2, where C1 if total C is below 10 and use C2 if C is between 10 and 15.

function plant_b()
    
    model = Model(GAMS.Optimizer)
    set_optimizer_attribute(model, "OptCR", 0)
    
    @variables(model, begin
        y1, Bin
        y2, Bin
        y3, Bin
        m_A >= 0
        m_C1 >= 0
        m_C2 >= 0
        m_B0 >= 0
        m_B1 >= 0
        m_B2 >= 0
        m_B3 >= 0
        
    end)
    @constraints(model, begin
        process_one, m_B1 == 0.9 * m_A
        B_total, m_B0 + m_B1 == m_B2 + m_B3 # buy + made = used by 2 + used by 3
        C_total, m_C1 + m_C2 == 0.82 * m_B2 + 0.95 * m_B3 # total amount produced = conversion by 2 + conversion by 3
        demand_C2, 0.82 * m_B2 <= 15 * y2 # mass of B in process 2 should be 0 if off; otherwise, C it produces should be lower than demand
        demand_C3, 0.95 * m_B3 <= 15 * y3 # same as above
        demand_A, m_A <= 16 # demand of A
        exclusivity, y2 + y3 == 1 # we either use process 2 or 3
        first_demand_C, m_C1 <= 10
        second_demand_C, m_C2 <= 5 # this way, C1 + C2 = 15, which is the total demand
        
        
    end)
    @objective(model, Max, 1800 * m_C1 + 1500 * m_C2 - (500 * m_A + 950 * m_B1 + (1000 * y1 + 250 * m_A) + (1500 * y2 + 400 * m_B2) + (2000 * y3 + 550 * m_B3)))
    
    print(model)
    optimize!(model)
    println("objective: ", objective_value(model))
    println("y1: ", value.(y1))
    println("y2: ", value.(y2))
    println("y3: ", value.(y3))
    println("m_A: ", value.(m_A))
    println("m_B0: ", value.(m_B1))                    
    println("m_B1: ", value.(m_B1))
    println("m_B2: ", value.(m_B2))
    println("m_B3: ", value.(m_B3))
    println("m_C1: ", value.(m_C1))
    println("m_C2: ", value.(m_C2))
    
end
```
